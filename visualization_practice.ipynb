```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('StudentsPerformance.csv')
```

```python
df.head(10)
```

```python
plt.figure(figsize=(10,8))
plt.hist(df['math score'], color='blue', bins=int(1000/5), edgecolor='black')
plt.title("Histogram of Math Score")
plt.xlabel("Math Scores")
plt.ylabel('Frequency')
```

```python
# it would be worth exploring the relationship between
# a parental level of education and math scores
```

```python
#Full Disclosure: Used the article 
#"Histograms and Density Plots in Python" to guide this

plt.figure(figsize=(10,10))

# List of six parental levels of education to plot

level_of_ed = ['some high school', 'high school', 'some college',
         "associate's degree", "bachelor's degree", "master's degree"]

# Iterate through the six levels
for level in level_of_ed:
    # Subset to the level
    subset = df[df['parental level of education'] == level]
    
    # Draw the density plot
    sns.distplot(subset['math score'], hist = False, kde = True,
                 kde_kws = {'linewidth': 3},
                 label = level)
    
# Plot formatting
plt.legend(prop={'size': 16}, title = 'Level of Education')
plt.title('Density Plot of Math Scores with Multiple Levels of Parental Education')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
```

```python
df['parental level of education'].value_counts().plot(kind='bar')
```
